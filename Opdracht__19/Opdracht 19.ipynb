```python
import seaborn as sns

from matplotlib import pyplot as plt

from sklearn.cluster import KMeans

from sklearn import metrics
from sklearn.metrics import pairwise_distances
```

```python
pg = sns.load_dataset("penguins")
pg = pg.dropna()

pg.head()
```

```python
sns.pairplot(pg)
```

```python
features = ['bill_length_mm', 'bill_depth_mm', 'body_mass_g']
km = KMeans(n_clusters=3, random_state = 42).fit(pg[features])
```

```python
pg["cluster"] = km.predict(pg[features])
pg.head()
```

```python
sns.pairplot(pg, hue="cluster")
```

Hierboven is de grafiek te zien met 3 clusters.

```python
metrics.silhouette_score(pg[features], km.labels_, metric='euclidean')
```

```python
km.cluster_centers_
```

```python
 table = pg.groupby(['species', 'cluster']).size().unstack('cluster', fill_value=0)
table
```
